# Background
The Department of Government Efficiency (DOGE) was established under President Trump to identify and eliminate wasteful spending, streamline operations, and improve service delivery across federal agencies. By reviewing and restructuring contracts, DOGE aims to generate measurable cost savings while maintaining or enhancing operational effectiveness.

# Problem Statement
Although DOGE’s interventions are designed to reduce federal expenditures, the actual impact on small and minority-owned businesses remains uncertain. **We need to assess whether DOGE’s contract reviews and terminations deliver equitable savings across all vendor categories or whether procedural biases disadvantage historically underserved entities.**

# Data Science Objectives
Using merged datasets from DOGE terminations, and SAM.gov, we will:
1. Clean and preprocess the data—imputing missing values and encoding flags for analysis;
2. Engineer key features including award value, total dollars obligated, potential award ceiling, and contract duration;
3. Explore relationships among these variables and the realized savings; and
4. Build a predictive model to quantify how process-level factors drive cost reductions and to detect any disparities in savings for small and minority-owned vendors.

We have all the relevant data for this business case therefore we are working with a **supervised learning and regression/classification problem.**

### Department Equity Index

*EDS = (Dollar_Share – Action_Share) / Action_Share  
Negative ⇒ under-funded, Positive ⇒ over-funded*

**Goal:** Rank departments by Equity Deviation Score: percent over/under–performance.


**Insight-prompt:** Focus equity remediation on the bottom quintile.

**Summation:** The EDS for the Peace Corps is 11.46, and for the Veterans Affairs Department it is 25.82. A higher EDS indicates that the proportion of dollars received by minority-owned businesses is significantly higher than their proportion of actions.

```python
# Compute Equity Deviation Score
epsilon = 1e-6  # Small value to prevent division by zero

df_cleaned['EDS_Small']    = (
    df_cleaned['Small_Dollars_Share'] - df_cleaned['Small_Actions_Share']
) / (df_cleaned['Small_Actions_Share'] + epsilon)

df_cleaned['EDS_Minority'] = (
    df_cleaned['Minority_Dollars_Share'] - df_cleaned['Minority_Actions_Share']
) / (df_cleaned['Minority_Actions_Share'] + epsilon)

# Top/worst offenders
eds = df_cleaned[['Department Name','EDS_Small','EDS_Minority']].set_index('Department Name')
eds_sorted = eds.sort_values('EDS_Minority')
print("Most Under-funding (Minority):\n", eds_sorted.head(5))
print("\nMost Over-funding (Minority):\n", eds_sorted.tail(5))

# Heatmap
fig = px.imshow(
    eds.sort_values('EDS_Minority', ascending=False),
    color_continuous_scale='RdBu', aspect='auto',
    labels=dict(x="", y="Department", color="EDS"),
    title="Equity Deviation Score Heatmap"
)
fig.update_xaxes(side="top")
fig.show()
```

### Spending Reduction Simulation (10%)

**Goal:** Contrast
- Uniform 10% cut
- Large-biz–only cut
and their effects on small & minority spend


**Functionality:** In essence, this code simulates two different budget cut scenarios and visualizes their potential impact on small business spending across the top spending government departments.

**Insight-prompt:** Large-only cuts preserve small‐biz spend; uniform cuts harm them equally.

**Summation:** this code provides a visual comparison of how different budget reduction strategies (uniform vs. large-business-only cuts) might impact the amount of money allocated to small businesses by the departments with the highest small business spending.

```python
# Sim functions
def uniform_cut(df, rate=0.1):
    df2 = df.copy()
    for col in ['Total Dollars','Small Business Dollars','Total Minority-Owned Business Dollars']:
        df2[f'{col}_UniSim'] = df2[col] * (1-rate)
    return df2

def large_only_cut(df, rate=0.1):
    df2 = df.copy()
    df2['Total Dollars_LgSim'] = (
        df2['Total Dollars'] - df2['Large Business Dollars'] * rate
    )
    df2['Small Business Dollars_LgSim'] = df2['Small Business Dollars']  # unchanged
    df2['Total Minority-Owned Business Dollars_LgSim'] = df2['Total Minority-Owned Business Dollars']
    return df2

df_uni  = uniform_cut(df_cleaned)
df_lg   = large_only_cut(df_cleaned)

# Build comparison DF
cmp = df_cleaned[['Department Name','Small Business Dollars','Total Minority-Owned Business Dollars']].copy()
cmp['Small_UniSim']    = df_uni['Small Business Dollars_UniSim']
cmp['Minority_UniSim'] = df_uni['Total Minority-Owned Business Dollars_UniSim']
cmp['Small_LgSim']     = df_lg['Small Business Dollars_LgSim']
cmp['Minority_LgSim']  = df_lg['Total Minority-Owned Business Dollars_LgSim']

# Plot for Top 8 spenders
top8 = cmp.nlargest(8,'Small Business Dollars')
fig = go.Figure()
for scenario in ['Small Business Dollars','Small_UniSim','Small_LgSim']:
    fig.add_trace(go.Bar(
        y=top8['Department Name'], x=top8[scenario],
        name=scenario
    ))
fig.update_layout(
    barmode='group',
    title="Small Biz Spend: Before vs. Uniform & Large-Only Cuts",
    xaxis_title="Dollars",
    yaxis_title="",
    xaxis_range=[0, top8[['Small Business Dollars','Small_UniSim','Small_LgSim']].values.max() * 1.1] # Set x-axis range dynamically
)
fig.show()
```

###Outlier & Risk Detection (Isolation Forest Model)

**Goal:** Surface departments with anomalous minority‐spend‐per‐action.

**Functionality:** This code reveals hidden imbalances and red flags by highlighting anomalies in Minority-Owned Dollars per Action.
 Identify departments with:
1.  High actions but minimal dollars (or vice versa)
2.  Negative dollar figures (as seen in some rows)
3.  High concentrations of one vendor type
4.  Visualization:Interactive Bubble chart with axis = Actions vs. Dollars

**Insight-prompt:** Investigate these outliers for procedural quirks or data errors.


**Sumation:** this code calculates the average spend per action for minority businesses for each department, identifies departments with unusually high or low spend-per-action values using an anomaly detection algorithm, visualizes these anomalies on a scatter plot, and then lists the departments identified as outliers based on their spend-per-action.

```python
# Compute spend‐per‐action
df_cleaned['Min_Spend_per_Action'] = (
    df_cleaned['Total Minority-Owned Business Dollars'] /
    (df_cleaned['Total Minority-Owned Business Actions'] + 1)
)

# Unsupervised anomaly detection
iso = IsolationForest(contamination=0.05, random_state=42)
df_cleaned['Anomaly'] = iso.fit_predict(
    df_cleaned[['Min_Spend_per_Action']]
) == -1

# Interactive outlier plot
fig = px.scatter(
    df_cleaned, x='Total Minority-Owned Business Actions', y='Total Minority-Owned Business Dollars',
    color='Anomaly', symbol='Anomaly',
    hover_name='Department Name',
    title='Anomalous Minority Spend vs Actions'
)
fig.show()

# List outliers
outliers = df_cleaned[df_cleaned['Anomaly']].sort_values('Min_Spend_per_Action')
outliers[['Department Name','Min_Spend_per_Action']].head(10)
```

Data Cleaning

**To use model-based imputation with scikit-learn’s IterativeImputer:** We are essentially building mini regression models to predict missing values using the other features within the dataset. It’s like giving your data a second chance to fill in the blanks intelligently.

**How It Works:**
Each feature with missing values is modeled as a function of the other features in a round-robin fashion. It iteratively refines the estimates until convergence or a max number of iterations.

```python
# Handle and clean 'UNAVAILABLE', 'Unavailable', 'UNAVAILBE', 'UNAV' values in dataset

# Replace all "unavailable"-type strings with np.nan
unavailable_values = ['UNAVAILABLE', 'Unavailable', 'UNAVAILBE', 'UNAV']
df_doge.replace(unavailable_values, np.nan, inplace=True)

# Identify and drop columns with too much missing data (>50%), excluding 'piid'
threshold = len(df_doge) * 0.5
cols_to_drop = [col for col in df_doge.columns if col != 'piid' and df_doge[col].isnull().sum() > threshold]
df_doge_processed = df_doge.drop(columns=cols_to_drop)

# Separate numeric and non-numeric columns from the processed dataframe
numeric_cols = df_doge_processed.select_dtypes(include=[np.number]).columns
non_numeric_cols = df_doge_processed.select_dtypes(exclude=[np.number]).columns

# Model-based imputation for numeric columns
# Ensure that if numeric_cols is empty, we handle it
if len(numeric_cols) > 0:
    imputer = IterativeImputer(estimator=RandomForestRegressor(), max_iter=10, random_state=0)
    numeric_imputed = imputer.fit_transform(df_doge_processed[numeric_cols])
    df_numeric_imputed = pd.DataFrame(numeric_imputed, columns=numeric_cols, index=df_doge_processed.index)
else:
    df_numeric_imputed = pd.DataFrame(index=df_doge_processed.index) # Create an empty dataframe with the same index


# Fill missing values in non-numeric columns (e.g., with mode or 'Unknown')
df_non_numeric = df_doge_processed[non_numeric_cols].copy()
for col in df_non_numeric.columns:
    df_non_numeric[col].fillna('Unknown', inplace=True)


# Combine the imputed numeric and non-numeric data
# Ensure 'piid' is included if it exists in the processed dataframe
if 'piid' in df_doge_processed.columns:
     df_doge_cleaned = pd.concat([df_numeric_imputed, df_non_numeric], axis=1)
     # Ensure 'piid' is the first column for clarity if it exists
     if 'piid' in df_doge_cleaned.columns:
         cols = ['piid'] + [col for col in df_doge_cleaned.columns if col != 'piid']
         df_doge_cleaned = df_doge_cleaned[cols]
else:
    # If 'piid' was not in the processed dataframe (e.g., all were NaN), just combine numeric and non-numeric
    df_doge_cleaned = pd.concat([df_numeric_imputed, df_non_numeric], axis=1)
```

### T-Test & Wilcox Test (Rankings) for Statistical Significance

```python
# Separate savings for small/minority businesses and other businesses from terminated contracts
savings_small_minority = df_terminated[df_terminated['is_small_or_minority'] == True]['savings'].dropna()
savings_other = df_terminated[df_terminated['is_small_or_minority'] == False]['savings'].dropna()

# Perform an independent samples t-test (assuming normality and equal variances for now)
# A non-parametric test like Wilcoxon rank-sum test could be used if assumptions are not met
t_stat, p_value = stats.ttest_ind(savings_small_minority, savings_other)

print(f"Independent Samples t-test results:")
print(f"  T-statistic: {t_stat:.4f}")
print(f"  P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nInterpretation: The difference in mean savings between the two groups is statistically significant.")
else:
    print("\nInterpretation: The difference in mean savings between the two groups is not statistically significant.")

# Consider also performing a non-parametric test (Wilcoxon rank-sum test) as savings data is often not normally distributed
wilcoxon_stat, wilcoxon_p_value = stats.ranksums(savings_small_minority, savings_other)

print(f"\nWilcoxon Rank-Sum Test results:")
print(f"  Statistic: {wilcoxon_stat:.4f}")
print(f"  P-value: {wilcoxon_p_value:.4f}")

# Interpret the results
if wilcoxon_p_value < alpha:
    print("\nInterpretation: The difference in the distributions of savings between the two groups is statistically significant.")
else:
    print("\nInterpretation: The difference in the distributions of savings between the two groups is not statistically significant.")
```
